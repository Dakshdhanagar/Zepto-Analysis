```python
## handling missing values in sales data
```

```python
print(df_sales.isnull().sum())
```

```python
df_sales= df_sales.dropna(subset=['city','delivery_status'])
df_sales.isnull().sum()
```

```python
df_salles_mean=df_sales["delivery_time_mins"].mean()
df_sales["delivery_time_mins"].fillna(df_salles_mean,inplace=True)
print(df_sales["delivery_time_mins"].isnull().sum())
df_sales.isnull().sum()
```

```python
# Duplicates recods in sales data

df_sales.duplicated().sum()
```

```python
df_sales.drop_duplicates(inplace=True)
df_sales.duplicated().sum()
```

```python
# Formatting the orderDate column to datetime format
df_sales['order_date']=pd.to_datetime(df_sales['order_date'])
print(df_sales['order_date'])
```

```python
# Find min, max, and average total amount
min_total_amount = df_sales['total_amount'].min()
max_total_amount = df_sales['total_amount'].max()   
avg_total_amount = df_sales['total_amount'].mean()
print(f"Minimum Total Amount: {min_total_amount}")      
print(f"Maximum Total Amount: {max_total_amount}")
print(f"Average Total Amount: {avg_total_amount}")
```

```python
# Top 5 products by total sales amount
q=df_sales["product_id"].duplicated().sum()
print(q)
top_products= df_sales.groupby('product_id')['total_amount'].sum().nlargest(5)
print("Top 5 Products by Total Sales Amount:",top_products)
```

```python
# Merge with product details to get product names
# top_products.reset_index()
top_products_details = top_products.reset_index().merge(
    df_products,
    on="product_id"
)
print("Top 5 Products with Details:", top_products_details)
```

```python
# Total sales by city
sales_by_city=df_sales.groupby('city')['total_amount'].sum().sort_values(ascending=False)
print(sales_by_city)
```

```python
#  Average delivery time by city
avg_delivery_time_by_city = df_sales.groupby('city')['delivery_time_mins'].mean().sort_values(ascending=False)
print(avg_delivery_time_by_city)
```

```python
# Sales trend over time (e.g., monthly sales)
df_sales['month'] = df_sales['order_date'].dt.to_period('M')
monthly = df_sales.groupby('month')['total_amount'].sum()
print(monthly)
```

```python
df_sales.columns
```

```python
# Sales by product category
sales_by_category = df_sales.merge(df_products, on='product_id').groupby('category')['total_amount'].sum().sort_values(ascending=False)
print(sales_by_category)
```

```python
# Relationship between city and total sales amount



sales_by_city = df_sales.groupby("city")["total_amount"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=sales_by_city.index, y=sales_by_city.values,palette="viridis")
plt.title("Total Sales by City")
plt.xlabel("City")
plt.ylabel("Total Sales Amount")
# plt.xticks(rotation=45)
# plt.tight_layout()
plt.show()

# plt.figure(figsize=(10,6))
# sns.lineplot(x='city', y='total_amount', data=df_sales)
# plt.title("City vs Total Sales")
# plt.xlabel("City")
# plt.ylabel("Total Amount")
# plt.show()
```

```python
# relationship between city Vs Average Delivery Time

avg_delivery = df_sales.groupby("city")["delivery_time_mins"].mean().sort_values()

plt.figure(figsize=(12,12))
sns.barplot(x=avg_delivery.index, y=avg_delivery.values,palette="Set2")
plt.ylim(24.5, 26.5)
plt.title("Average Delivery Time by City")
plt.xlabel("City")
plt.ylabel("Average Delivery Time (Minutes)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Relationship between month and total sales amount

# df_sales["month"] = df_sales["order_date"].dt.to_period("M")
monthly_sales = df_sales.groupby("month")["total_amount"].sum()

plt.figure(figsize=(14,7))
monthly_sales.plot(marker="o",color="purple")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales Amount")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Relationship between Category Vs Total Sales Amount
sales_category = df_sales.merge(df_products, on="product_id")
sales_category = sales_category.groupby("category")["total_amount"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,8))

plt.pie(
    sales_category.values,
    labels=sales_category.index,
    autopct="%1.1f%%",
    startangle=140
)

plt.title("Total Sales by Product Category")
plt.axis("equal")   # circle proper dikhe
plt.tight_layout()
plt.show()
```
